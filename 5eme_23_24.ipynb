## 3.1 - Expression littérale (simplification d'écriture)

### 3.1.0 - Définition de l'expression littérale

<p><strong>Une expression littérale est une expression contenant une ou plusieurs lettres, ces lettres désignent des nombres.</strong></p>

### 3.1.1 Exemples :

<p><ol>
    <li>Périmètre d'un carré de côté a : <em>P = a + a + a + a = 4 x a</em></li>
    <li>Aire d'un carré de côté a : <em>A = a x a = a²</em></li>
    <li>Périmètre d'un rectangle de longueur a et largeur b : <em>P = a + b + a + b = 2 x a + 2 x b</em></li>
    <li>Aire d'un rectangle de longueur a et largeur b : <em>A = a x b</em></li>
    <li>Périmètre d'un triangle de côté a, b et c : <em>P = a + b + c</em></li>
    <li>Périmètre d'un triangle de base b et de hauteur h : <em>A = b * h / 2</em></li>
</ol></p>

### 3.1.2 - Simplification d'écriture

#### Convention d'écriture 1
<p>Le signe "x" de la multiplication peut être supprimé (ou remplacé par un point) devant une lettre ou une parenthèse.</p>

#### Exemples :
<p><ol>
    <li><em>4 x a = 4a </em>ou bien<em> 4 x a = 4.a</em></li>
    <li><em>4 x (a + 2) = 4(a + 2) </em>ou bien<em> 4 x (a + 2) = 4.(a + 2)</em></li>
    <li><em>a x b = ab </em>ou bien<em> a x b = a.b</em></li>
    <li><em>a x (a + 4) = a(a + 4) </em>ou bien<em> a x (a + 4) = a.(a+ 4)</em></li>
    <li><em>(4 + a) x (2 + b) = (4 + a)(2 + b)</em> ou bien <em> (4 + a) x (2 + b) = (4 + a).(2 + b)</em></li>
    <li><em>4 x a x b = 4ab </em>ou bien<em> 4 x a x b = 4.a.b</em></li>
    <li><em>4 x a x (2 + b) x c = 4ac(2 + b)</em> ou bien <em> 4 x a x (2 + b) x c = 4.a.c.(2 + b)</em></li>
</ol></p>

#### Convention d'écriture 2
<p><ol>
    <li><em>a x a = a² </em>&nbsp;se lit "a au carré"</li>
    <li><em>a x a x a = a³ </em>&nbsp;se lit "a au cube"</li>
    <li><em>a x a x a x a = a⁴ </em>&nbsp;se lit "a à la puissance 4"</li>
</ol></p>

## 2 - Tester une égalité

### 3.2.1 Appliquer une formule

#### Méthode

<p>Calculer la valeur d'une expression littérale, c'est attribuer un nombre à chaque lettre afin d'évaluer l'expression.</p>

#### Exemples 

<p><ol>
    <li>Périmètre d'un carré de côté a pour a = 2 : <em>P = a + a + a + a = 4 x a</em></li>
    <li>Aire d'un carré de côté a pour a = 2 : <em>A = a x a = a²</em></li>
    <li>Périmètre d'un rectangle de longueur a et largeur b pour a = 2 et b = 3 : <em>P = a + b + a + b = 2 x a + 2 x b</em></li>
    <li>Aire d'un rectangle de longueur a et largeur b a = 2 et b = 3 : <em>A = a x b</em></li>
    <li>Périmètre d'un triangle de côté a, b et c a = 2 et b = 3 et c = 5 : <em>P = a + b + c</em></li>
    <li>Périmètre d'un triangle de base b et de hauteur h pour b = 3 et h = 4 : <em>A = b * h / 2</em></li>
</ol></p>

### 3.2.2 Tester une égalité

#### Définition

<p>En mathématiques, l’égalité est une relation entre deux expressions signifiant que ces expressions sont identiques, c’est-à-dire que le remplacement de l’un par l’autre dans une expression ne change jamais la valeur de cette dernière.</p>

#### Méthode

<p>Pour tester si une égalité est vraie pour des valeurs numériques attribuées aux lettres :</p>
<ol>
    <li>On calcule la valeur du <strong>membre de gauche</strong> en remplaçantchaque lettre par le nombre donné.</li>
    <li>On calcule la valeur du <strong>membre de droite</strong> en remplaçantchaque lettre par le nombre donné.</li>
    <li>On observe l'égalité ou non des deux valeurs obtenues et on conclut.</li>
</ol>

# Chapitre 4 - Utiliser une symétrie

## 4.1 - Symétrie axiale (rappel)

### 4.1.1 Definition 

<p>
    Deux figures sont symétriques par rapport à une
droite si elle se superposent par pliage le long de la droite.
</p>

### 4.1.2 Propriété

<p>
    Le point M' est symétrique à M par rapport à la droite (d) signifie :
</p>

<ul>
    <li>(MM') est perpendiculaire à (d).</li>
    <li>M et M' sont à égale distance de (d)</li>
</ul>
    
### 4.1.3 Remarque

<p>
    Si M' est le symétrique de M par rapport à (d) alors (d) est la médiatrice de [MM'].
</p>
